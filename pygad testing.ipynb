```python
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
data = pd.read_csv('mushrooms.csv')
# get y values
yvals = []
for d in data.iterrows():
    if d[1]['class'] == 'p':
        yvals.append(1)
    else:
        yvals.append(0)

# drop x column
data.drop(columns=data.columns[0], axis=1, inplace=True)
```

```python
# fill missing data with mode
for column in data.columns:
    data[column].fillna(data[column].mode()[0], inplace=True)
```

```python
# convert to numerical attributes
dataframes = []
for column in data.columns:
    o_h_encoded = pd.get_dummies(data[column], prefix=str(column))
    dataframes.append(o_h_encoded)
cleaned_data = pd.concat(dataframes, axis=1)
```

```python
# get xvals
xvals = []
for d in cleaned_data.iterrows():
    info = d[1]
    xvals.append(list(info))
```

```python
from sklearn.metrics import log_loss, zero_one_loss
from scipy.special import expit
def zero_one(true_labels, pred_labels): 
    return zero_one_loss(true_labels, np.rint(pred_labels).astype(int), normalize=True)

def get_utility(candidate, _):
    alpha = candidate[:-1]
    beta = candidate[-1]
    predicitons = [expit(np.dot(x, alpha) + beta) for x in xvals]
    loss = zero_one(yvals, predicitons)
    return loss
```

```python
import random
import math
def mutation_func(offspring, ga_instance):
    for chromosome_idx in range(offspring.shape[0]):
        random_gene_idx = numpy.random.choice(range(offspring.shape[0]))

        offspring[chromosome_idx, random_gene_idx] += numpy.random.normal(scale=math.sqrt(0.1))

    return offspring
```

```python
# crossover probabiltiy isn't working for some reason
# this is also a lot less effective than the default params
import pygad 
import numpy

def on_start(ga_instance): print("on_start()")

def on_fitness(ga_instance, population_fitness): 
    print(max(population_fitness))

def on_stop(ga_instance, last_population_fitness): 
    print("on_stop()")

ga_instance = pygad.GA(num_generations=100, 
                       num_parents_mating=100, 
                       fitness_func=get_utility, 
                       sol_per_pop=100, 
                       num_genes=len(xvals[0])+1, 
                       parent_selection_type="rws", 
                       crossover_type="uniform",
                       mutation_type=mutation_func,
                       mutation_probability=1,
                       on_start=on_start, 
                       on_fitness=on_fitness, 
                       on_stop=on_stop)

ga_instance.run()
```
